## **Data Cleaning**
Points covered under this section:<br/>
*• Selecting wanted columns*<br/>
*• Removing null values*<br/>
*• Calculate current run rate*<br/>
*• Creating Target column*<br/>

```python
data = df[[
    'match_id',
    'innings',
    'batting_team',
    'bowling_team',
    'over',
    'ball',
    'team_runs',
    'team_wicket'
]]
```

```python
data.isnull().sum()
```

```python
data.duplicated().sum()
```

```python
data = data.copy()

data.drop_duplicates(inplace=True)
```

```python
data['current_over'] = data["over"] + (data["ball"]/10)
```

```python
data["balls_left"] = 120 - (data["over"] * 6) + data["ball"]
```

```python
data["crr"] = data["team_runs"] / data["current_over"]
```

```python
data['crr'] = data['crr'].replace([np.inf, -np.inf], 0)
```

```python
final_score = data.groupby(["match_id", "innings"])["team_runs"].max().reset_index()
```

```python
final_score.rename(columns={"team_runs":"final_score"}, inplace=True)
```

```python
data = data.merge(final_score, on=["match_id","innings"])
```

```python
corr_matrix = data.corr(numeric_only=True)
top_corr_features = corr_matrix.index

plt.figure(figsize=(13,10))

sns.heatmap(
    data=data[top_corr_features].corr(numeric_only=True),
    annot=True,
    cmap="RdYlGn"
)


plt.show()
```

## **Data Preprocessing**
*• Handling categorical features*<br/>
*• Splitting dataset into train and test set on the basis of date*

```python
data = pd.get_dummies(
    data,
    columns=["batting_team", "bowling_team"]
)
```

```python
data = data.astype(int)
data
```

```python
x = data.drop(["match_id","final_score"], axis=1)
y = data["final_score"]
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test= train_test_split(x,y, test_size=0.4, random_state=42)
```

## Linear Regression 

```python
#Linear Regression model
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred_lr = lr.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error, r2_score

print("RMSE:", np.sqrt(mean_absolute_error(y_test, y_pred_lr)))
print("MAE:", mean_absolute_error(y_test, y_pred_lr))
print("r2_score:", r2_score(y_test, y_pred_lr))
```

## Decision Tree 

```python
#Decision Tree Model
from sklearn.tree import DecisionTreeRegressor

dt = DecisionTreeRegressor()
dt.fit(x_train, y_train)
```

```python
y_pred_dt = dt.predict(x_test)
```

```python
print("RMSE:", np.sqrt(mean_absolute_error(y_test, y_pred_lr)))
print("MAE:", mean_absolute_error(y_test, y_pred_lr))
print("r2_score:", r2_score(y_test, y_pred_lr))
```

## Random Forest 

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor()
rf.fit(x_train, y_train)
```

```python
y_pred_rf = rf.predict(x_test)
```

```python
rmse = np.sqrt(mean_absolute_error(y_test, y_pred_rf))
mae = mean_absolute_error(y_test, y_pred_rf)
r2 =  r2_score(y_test, y_pred_rf)
```

```python
print("RMSE:", rmse)
print("MAE:", mae)
print("r2_score:", r2)
```

## XGBoost

```python
#XGBoost Model
from xgboost import XGBRegressor

xgb = XGBRegressor(
    n_estimators=200,
    learning_rate=0.05,
    max_depth=6,
    random_state=42
)
xgb.fit(x_train, y_train)
```

```python
y_pred_xgb = xgb.predict(x_test)
```

```python
xgb_rmse = np.sqrt(mean_absolute_error(y_test, y_pred_xgb))
xgb_mae = mean_absolute_error(y_test, y_pred_xgb)
xgb_r2 =  r2_score(y_test, y_pred_xgb)
```

```python
print("RMSE:", xgb_rmse)
print("MAE:", xgb_mae)
print("r2_score:", xgb_r2)
```

### CatBoost

```python
#CatBoost Model
from catboost import CatBoostRegressor

cat = CatBoostRegressor()
cat.fit(x_train,y_train)
```

```python
cat_rmse = np.sqrt(mean_absolute_error(y_test, y_pred_cat))
cat_mae = mean_absolute_error(y_test, y_pred_cat)
cat_r2 =  r2_score(y_test, y_pred_cat)
```

```python
print("RMSE:", cat_rmse)
print("MAE:", cat_mae)
print("r2_score:", cat_r2)
```

### Hyperparameter Tuning

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
params = {
    'n_estimators': [100,200,300],
    'learning_rate': [0.01,0.05,0.1],
    'max_depth':[4,6,8],
    'subsample':[0.8,1.0]
}
```

```python
random_search = RandomizedSearchCV(
    estimator=XGBRegressor(random_state=42),
    param_distributions=params,
    n_iter=10,
    cv=3,
    verbose=2,
    n_jobs=-1
)
```

```python
random_search.fit(x_train,y_train)
```

```python
random_search.best_params_
```

### Residual Analysis

```python
residuals = y_test - y_pred_xgb
```

```python
best_pred = best_xgb.predict(x_test)
```

```python
rs_rmse = np.sqrt(mean_absolute_error(y_test, best_pred))
rs_mae = mean_absolute_error(y_test, best_pred)
rs_r2 =  r2_score(y_test, best_pred)
```

```python
print("RMSE:", rs_rmse)
print("MAE:", rs_mae)
print("r2_score:", rs_r2)
```

```python
plt.figure(figsize=(8,6))

sns.histplot(residuals, kde=True)

plt.title("Residual Distribution")

plt.show()
```
